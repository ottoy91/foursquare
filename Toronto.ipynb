```python
import numpy as np

import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
#Link to postal codes in Wikipedia
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

tables = pd.read_html(link,header=0)

#Set headers
headings = ['PostalCode', 'Borough', 'Neighborhood']
for table in tables:
    current_headings = table.columns.values[:2]
    if current_headings == headings:
        break
        
```

```python
#Turn first html column to data frame
hoods = pd.DataFrame(tables[0])
hoods.head()
```

```python
#Drop Bourough = 'Not assigned' form Data Frame
hoods = hoods[hoods.Borough != 'Not assigned']
hoods.head()
```

```python
#Data source has been updated: no neighborhoods without assignment exist where borough is assigned
hoods[hoods.Neighborhood == 'Not assigned']
```

```python
#Group by postal code such that Neighborhoods are separated by comma
hoods= hoods.groupby(['Postal Code', 'Borough']).Neighborhood.agg([('Neighborhood', ', '.join)]).reset_index()
hoods
```

```python
#join Boroughs under one postal code
hoods = hoods.groupby('Borough').agg({'Postal Code':'first', 
                             'Neighborhood': ', '.join 
                              }).reset_index()
hoods
```
